```python
import pandas as pd
import numpy as np

dfventas = pd.read_csv('ventas.csv')

# 1. Cantidad total por producto
ventas_producto_categoria = dfventas.groupby('Producto')['Cantidad'].sum().reset_index()
print("\n Cantidad total vendida por producto:")
print(ventas_producto_categoria)

# 2. Producto con mayor ventas
producto_mas_vendido = ventas_producto_categoria.sort_values(by='Cantidad', ascending=False).iloc[0]
print(f"Producto con mayor cantidad vendida es el: {producto_mas_vendido['Producto']}")


# 3. Precio promedio
precio_promedio = dfventas['Precio_Unitario'].mean()
print("\n Precio promedio de los productos vendidos:", precio_promedio)
```

```python
import pandas as pd
import numpy as np

dfclima = pd.read_csv('clima.csv')

# 1. Temperatura promedio por ciudad
temp_promedio = dfclima.groupby('Ciudad')['Temperatura'].mean().reset_index()
print("Temperatura promedio por ciudad:\n", temp_promedio)

# 2. Registros con las temperaturas mas altas y mas bajas de todo el dataset
temp_max = dfclima.loc[dfclima['Temperatura'].idxmax()]
temp_min = dfclima.loc[dfclima['Temperatura'].idxmin()]
print(f"\n Registro de temperatura más alta: {temp_max['Temperatura']} grados")
print(f"\n Registro de temperatura más baja: {temp_min['Temperatura']} grados")

# 3. Ciudad con temperatura más alta y baja en todo el dataset
ciudad_masalta = dfclima.groupby('Ciudad')['Temperatura'].max().idxmax()
ciudad_masbaja = dfclima.groupby('Ciudad')['Temperatura'].min().idxmin()
print(f"\n Ciudad con la temperatura más alta: {ciudad_masalta}")
print(f"\n Ciudad con la temperatura más baja: {ciudad_masbaja}")

# 4. Registros con una temperatura >30°C
calor_extremo = dfclima[dfclima['Temperatura'] > 30].shape[0]
print(f"\n Días con temperatura >30°C: {calor_extremo}")

# 5. Días registrados por ciudad en total
dias_por_ciudad = dfclima['Ciudad'].value_counts().reset_index()
dias_por_ciudad.columns = ['Ciudad', 'Días']
print(f"\n Días registrados por ciudad:\n", dias_por_ciudad)
```

```python
import pandas as pd
import numpy as np

dfcalificaciones = pd.read_csv('calificaciones.csv')

# 2. Promedio por materia
promedio_materia = dfcalificaciones.groupby('Materia')['Calificación'].mean().reset_index()
print("\n Promedio de calificaciones por materia:\n", promedio_materia)

# 3. Encontrar al estudiante con mejor promedio
mejor_estudiante = dfcalificaciones.groupby('Estudiante')['Calificación'].mean().sort_values(ascending=False).head(1)
print(f"\nEstudiante con mejor promedio: {mejor_estudiante.index[0]}")


# 5. Estudiantes con promedio >85
promedios = dfcalificaciones.groupby('Estudiante')['Calificación'].mean()
mayor_85 = promedios[promedios > 85]
cantidad_mayor85 = mayor_85.count()
print(f"\nCantidad de estudiantes con promedio mayor a 85: {cantidad_mayor85}")


# 6. Materia con más calificaciones
materia_mas_calificada = dfcalificaciones['Materia'].value_counts().idxmax()
print("\n Materia con más calificaciones registradas:", materia_mas_calificada)


# 7. Estudiantes con peor promedio (los 5 más bajos)
promedios = dfcalificaciones.groupby('Estudiante')['Calificación'].mean()
promediosbajos_5 = promedios.sort_values(ascending=True).head(5). reset_index()
print("\n Los 5 estudiantes con el promedio más bajo:", promediosbajos_5)



```
